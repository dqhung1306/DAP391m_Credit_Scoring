```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os 

current_dir = os.getcwd()
parent_dir = os.path.dirname(current_dir)
file_path = os.path.join(parent_dir, 'data', 'raw_data', 'application_train.csv')

appl = pd.read_csv(file_path)
appl.head()
```

```python
# checking missing data
total = appl.isnull().sum().sort_values(ascending = False)
percent = (appl.isnull().sum()/appl.isnull().count()*100).sort_values(ascending = False)
missing  = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing[missing["Percent"] > 50]
```

* data is imbalanced or not

```python
target_in_percent = 100 * appl["TARGET"].value_counts()/len(appl)
target_in_percent
```

```python
plot_data = appl["TARGET"].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(plot_data, palette='viridis')
plt.title("Distribution of target")
plt.show()
```

```python
plot_data = appl['NAME_CONTRACT_TYPE'].value_counts()
plt.figure(figsize=(6, 4))
plt.pie(
    plot_data,
    labels=plot_data.index,
    autopct='%1.1f%%',
    startangle=90,            
    colors=['skyblue', 'red'] 
)
plt.title("Distribution by type of loans")
plt.show()
```

```python
plot_data = appl['CODE_GENDER'].value_counts()
plt.figure(figsize=(6, 4))
plt.pie(
    plot_data,
    labels=plot_data.index,
    autopct='%1.1f%%',        
    startangle=90
)
plt.title("Distribution by gender")
plt.show()
```

```python
amt_cols = ["AMT_CREDIT", "AMT_INCOME_TOTAL", "AMT_GOODS_PRICE", "AMT_ANNUITY"]

fig, axes = plt.subplots(2, 2, figsize=(20, 12))
ax = axes.flatten()

for i, col in enumerate(amt_cols):
    sns.distplot(appl[col].dropna(), kde=True, ax=ax[i])
    ax[i].set_title("Distribution of {col}")

plt.tight_layout()
plt.show()
```
